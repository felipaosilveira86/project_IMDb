```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
class CAT():
    def __init__(self):
        self.__industry = pd.read_csv('base_consolidada.csv', sep=';', header = 0) # todos os filmes da indústria
        self.__netflix = self.__industry[self.__industry['on_netflix'] == True] # todos os filmes do catálogo netflix
        self.__other = self.__industry[self.__industry['on_netflix'] == False] # todos os filmes fora do catálogo netflix
        self.__actors = pd.read_csv('base_atores.csv', sep=';', header = 0) # a base de atores
        
    def netflix_catalog(self):
        return self.__netflix.drop(columns=['on_netflix']) # retorna o catálogo netflix
    
    def industry_catalog(self): # retorna o catálogo da indústria como um todo
        return self.__industry
    
    def other_catalog(self): # retorna só os filmes fora do catálogo netflix
        return self.__other.drop(columns=['on_netflix'])
    
    @property
    def industry(self):
        return self.__industry
    
    @property
    def netflix(self):
        return self.__netflix
    
    @property
    def other(self):
        return self.__other
    
    @property
    def actors(self):
        return self.__actors
    
    # Aí aqui programamos métodos que entreguem todos os stories que queremos!!!
    
    #realiza validação dos itens
    def input_limits(self, scope, metric, subfilter,house_content, limit, visualize, search_filter):
        # scope deve estar em ['industry', 'netflix', 'other']
        if scope not in ['industry', 'netflix', 'other']:
            scope = 'industry'
        
        # search_filter deve estar em ['genre', 'actors', 'year', 'country', 'language', 'director']
        if search_filter not in ['genre','actors', 'year', 'country', 'language', 'director']:
            raise Exception('Please choose valid filter. Valid filters: genre, actors, year, country, language, director.')
        if metric not in ['avg_vote', 'votes']:
            metric = 'avg_vote'

        # subfilter deve estar entre [['allgenders', 'females', 'males'], ['0age', '18age', '30age', '45age']]
        if subfilter[0] not in ['allgenders', 'females', 'males']:
            subfilter[0] = 'allgenders'
        if subfilter[1] not in ['allages','0age', '18age', '30age', '45age']:
            subfilter[1] = 'allages'
            
        # house_content = True filtrará 'production_company' == 'Netflix'
        if house_content not in [True, False]:
            house_content = False
           
        # limit é discricionário do usuário
        if type(limit) != int and type(limit) != float:
            limit = 100
            
        # visualize diz respeito ao retorno de gráficos com a análise, ao invés do DataFrame
        if visualize not in [True, False]:
            visualize = False   
        
        return (scope, metric, subfilter, house_content, limit, visualize)
    
    def plot_double_bar(self, search_filter, x1 , y1, x2, y2):

        fig, axes = plt.subplots(2, 1, figsize=(12, 8), dpi=150, constrained_layout=False)
        fig.tight_layout()
        ax1 = plt.subplot(211)
        ax1.set_title(f'Top 10 - Quantidade de {search_filter}')
        ax1.bar(x1[:10] , y1[:10] , color = 'gray')
        ax1.tick_params(rotation = 30)
        for index, value in enumerate(y1[:10]):
            plt.text(index - 0.15 , value / 2, s= value, fontsize=13)
        ax2 = plt.subplot(212, ymargin = 0.4)
        plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.3)
        ax2.set_title(f'Top 10 - Nota média por {search_filter}')
        ax2.bar(x2[:10] , y2[:10] , color = 'lime')
        for index, value in enumerate(y2[:10]):
            plt.text(index - 0.15 , value / 2, s= value, fontsize=13)
        ax2.tick_params(rotation = 30)
        
    def organize_data(self, copied_data, search_filter, composed_metric):
        
        rank = copied_data[search_filter].str.split(',', expand = True)
        rank = pd.concat([rank, copied_data[composed_metric]], axis = 1)
        rank[0]=rank[0].str.strip()
        rank[1]=rank[1].str.strip()
        rank[2]=rank[2].str.strip()
        s1 = rank[[0, composed_metric]].groupby(0).sum()
        s2 = rank[[1, composed_metric]].groupby(1).sum()
        s3 = rank[[2, composed_metric]].groupby(2).sum()
        total_nt = pd.concat([s1,s2,s3])
        total_nt = total_nt.groupby(total_nt.index)[composed_metric].sum()

        rank_genre = pd.concat([rank[0],rank[1],rank[2]]).value_counts()

        rank_genre = pd.DataFrame([total_nt, rank_genre]).T
        rank_genre.columns = ['Total', 'count']

        rank_genre['mean'] = round(rank_genre['Total'] / rank_genre['count'],1)

        return rank_genre[['count','mean']]
    
    def scope_df(self, scope):
        if scope == 'industry':
            copied_data = self.industry.copy()
        elif scope =='netflix':
            copied_data = self.netflix.copy()
        elif scope == 'other':
            copied_data = self.other.copy()
        return copied_data
    
    def ranking(self, search_filter, scope='industry', metric = 'avg_vote', house_content = False,
                subfilter = ['allgenders', 'allages'], limit = 100, visualize = False):

        # PRIMEIRO LIMITANDO OS INPUTS 
        tuple_ = self.input_limits(scope, metric, subfilter, house_content, limit, visualize, search_filter)
        scope, metric, subfilter, house_content, limit, visualize = tuple_
        
        # agregando a métrica e o subfilter escolhido para fazer a filtragem
        composed_metric = subfilter[0] + '_' + subfilter[1] + '_' + metric
        
        #Criando DF do scopo
        copied_data = self.scope_df(scope) 
        
        # a depender da métrica escolhida, a operação do groupby é outra; aqui, delimito isso
        if metric == 'avg_vote':
            operation = 'mean'
        else:
            operation = 'sum'

        ## AGORA FAZENDO A ANÁLISE

        if search_filter == 'actors':
            copied_data = self.actors.copy()
            if metric == 'votes':
                actors1 = copied_data.groupby('main_actor')[composed_metric].sum()
                # agora para supporting 1
                actors2 = copied_data.groupby('supporting_actor_1')[composed_metric].sum().apply(lambda x: x /1.1)
                # agora para o supporting 2
                actors3 = copied_data.groupby('supporting_actor_2')[composed_metric].sum().apply(lambda x: x /1.3)
                # juntando
                all_actors = pd.concat([actors1, actors2, actors3])
                all_actors = all_actors.groupby(all_actors.index).sum().sort_values(ascending=False)
                return all_actors.head(limit).index
                # FALTAM OS GRÁFICOS

            elif metric == 'avg_vote':
                actors1 = copied_data.groupby('main_actor')[composed_metric].agg(['mean', 'count'])
                actors1 = actors1[actors1['count'] > 5].sort_values(by=['mean', 'count'], ascending=[False, False])['mean']
                # agora para supporting 1
                actors2 = copied_data.groupby('supporting_actor_1')[composed_metric].agg(['mean', 'count'])
                actors2 = actors2[actors2['count'] > 5].sort_values(by=['mean', 'count'], ascending=[False, False])['mean']
                # agora para supporting 2
                actors3 = copied_data.groupby('supporting_actor_2')[composed_metric].agg(['mean', 'count'])
                actors3 = actors3[actors3['count'] > 5].sort_values(by=['mean', 'count'], ascending=[False, False])['mean']
                # juntando
                all_actors = pd.concat([actors1, actors2, actors3])
                return all_actors.groupby(all_actors.index).mean().sort_values(ascending=False).head(limit).index
                # FALTAM OS GRÁFICOS
            
        elif scope == 'industry':
            copied_data = self.__industry.copy()
            output = copied_data.groupby(search_filter)[composed_metric].agg([operation, 'count'])
            if visualize == False:
                return output.sort_values(by = [operation, 'count'], ascending=[False, False])[operation].head(limit)
            else:
                # CODAR AQUI OS GRÁFICOS
                pass
                
        elif scope == 'netflix':
            if house_content == True:
                copied_data = self.__netflix[self.netflix['production_company'] == 'Netflix'].copy()
                output = copied_data.groupby(search_filter)[composed_metric].agg([operation, 'count'])
            else:
                copied_data = self.__netflix.copy()
                output = copied_data.groupby(search_filter)[composed_metric].agg([operation, 'count'])
            if visualize == False:
                return output.sort_values(by = [operation, 'count'], ascending=[False, False])[operation].head(limit)
            else:
                # CODAR AQUI OS GRÁFICOS
                pass
        else:
            copied_data = self.__other.copy()
            output = copied_data.groupby(search_filter)[composed_metric].agg([operation, 'count'])
            if visualize == False:
                return output.sort_values(by = [operation, 'count'], ascending=[False, False])[operation].head(limit)
            else:
                # CODAR AQUI OS GRÁFICOS
                pass
        
        # genre, country and director
        if search_filter in ['genre', 'country', 'language']:
            
            if visualize == False:
                df_ = self.organize_data(copied_data, search_filter, composed_metric)
                print(df_.sort_values(by = 'count', ascending = False))

            else:
                df_ = self.organize_data(copied_data, search_filter, composed_metric)
                df1 = df_.sort_values(by = 'count', ascending = False)
                df2 = df_.sort_values(by = 'mean', ascending = False)
                #chamando a função para plotar o gráfico
                self.plot_double_bar(search_filter = search_filter, x1 = df1.index, y1 = df1['count'], x2 = df2.index, y2 = df2['mean'])
        
        # year e director
        elif search_filter in ['year', 'director']:
            copied_data.year = copied_data.year.apply(lambda x: str(x))
            df1 = copied_data[search_filter].value_counts().sort_values(ascending = False)
            df2 = copied_data[[search_filter, composed_metric]].groupby(search_filter).mean().sort_values(by = composed_metric, ascending = False)
            
            # df2.search_filter = df2.search_filter.apply(lambda x: str(x))
            if visualize == False:
                return df1
            else:
                self.plot_double_bar(search_filter = search_filter, x1 = df1.index, y1 = df1.values, x2 = df2.index, y2 = df2[composed_metric])
```

```python
program = CAT()
```

```python
program.ranking('year', visualize = True)
```
